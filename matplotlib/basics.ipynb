```python
import matplotlib.pyplot as plt
```

```python
# Plot a line graph with x and y values
plt.plot([1, 2, 3], [4, 5, 6], label="Line1")

# title to the plot
plt.title('Simple line plot')

# label for the x-axis
plt.xlabel('X axis')

# label for the y-axis
plt.ylabel('Y axis')

# legend to the plot
plt.legend()

# Enable grid lines on the plot
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Bar plot

# Create a bar plot with categories 'A', 'B', 'C' and their corresponding values 1, 2, 3
plt.bar(['A', 'B', 'C'], [1, 2, 3])

# Add a title to the bar plot
plt.title('Bar Plot')

# Label the x-axis
plt.xlabel('X axis')

# Label the y-axis
plt.ylabel('Y axis')

# Display the bar plot
plt.show()
```

```python
# Scatter plot

# Create a scatter plot with x and y values
plt.scatter([1, 2, 3], [4, 5, 6]) # (1,4), (2,5), (3,6)

# Add a title to the scatter plot
plt.title('Scatter Plot')

# Label the x-axis
plt.xlabel('X axis')

# Label the y-axis
plt.ylabel('Y axis')

# Display the scatter plot
plt.show()
```

```python
# Histogram

# Create a histogram with the given data
# The data represents the values and their frequencies
plt.hist([1, 2, 2, 2, 3, 3, 4, 4, 4, 4])

# Add a title to the histogram
plt.title('Histogram')

# Label the x-axis to represent the values
plt.xlabel('Value')

# Label the y-axis to represent the frequency of the values
plt.ylabel('Frequency')

# Display the histogram
plt.show()
```

```python
# Pie chart
# Define the data and labels for the pie chart
data = [10, 20, 30]
labels = ['A', 'B', 'C']

# Create a pie chart with the data and labels
# 'autopct' is used to display the percentage on the chart
plt.pie(data, labels=labels, autopct='%1.1f%%')

# Add a title to the pie chart
plt.title('Pie Chart')

# Display the pie chart
plt.show()
```

```python
# Box plot

# Create a box plot with three sets of data
# The data represents three groups with their respective values
plt.boxplot([[7, 8, 9], [5, 6, 7], [3, 4, 5]])

# Add a title to the box plot
plt.title('Box Plot')

# Display the box plot
plt.show()
```

```python
# Create a figure and a set of subplots with 1 row and 2 columns
fig, axs = plt.subplots(1, 2)

# Plot a line graph on the first subplot (axs[0])
axs[0].plot([1, 2, 3], [4, 5, 6])       # Line plot with points (1,4), (2,5), (3,6)

# Create a bar plot on the second subplot (axs[1])
axs[1].bar(['A', 'B', 'C'], [1, 3, 2])  # Bar plot with categories 'A', 'B', 'C' and values 1, 3, 2

# Display the plots
plt.show()
```
